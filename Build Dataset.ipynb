```python
import pandas as pd
import numpy as np
import synthetic
```

```python
data_files = [pd.read_csv(f) for f in ["data1.csv","data2.csv","data3.csv","data4.csv"]]
swap_set = pd.read_csv("data5.csv").astype(np.str)
```

```python
full_data = pd.concat(data_files).astype(np.str)
```

```python
reduced_data = full_data.drop(["first_name","first_name2","last_name","last_name2","email","address","city2","zip2","state2"],axis=1)
```

```python
reduced_data.columns = ["first_name","last_name","email","address","ssn","sex","city","zip","state","dob","phone","phone2","phone3","race","pcp_npi","suffix","title"]
```

```python
left = reduced_data.copy()
right = reduced_data.copy()
```

```python
synthetic.string_transpose(left,4,0.05)
print("Transpose Complete.")
synthetic.string_delete(left,3,0.05)
print("Delete Complete.")
synthetic.string_insert(left,3,0.05)
print("Insert Complete.")
synthetic.soundex_string_corrupt(left,0.1,columns=["first_name","last_name"])
print("Soundex Corrupt Complete.")
synthetic.edit_values(left,swap_set,0.1)
print("Edit Values Complete.")
synthetic.drop_per_column(left,drop_pct=0.08)
print("Per-Column Drop Complete.")
```

```python
synthetic.string_transpose(right,4,0.05)
print("Transpose Complete.")
synthetic.string_delete(right,3,0.05)
print("Delete Complete.")
synthetic.string_insert(right,3,0.05)
print("Insert Complete.")
synthetic.soundex_string_corrupt(right,0.1,columns=["first_name","last_name"])
print("Soundex Corrupt Complete.")
synthetic.edit_values(right,swap_set,0.1)
print("Edit Values Complete.")
synthetic.drop_per_column(right,drop_pct=0.08)
print("Per-Column Drop Complete.")
```

```python
left.to_csv("corrupted_left.csv",index=None)
right.to_csv("corrupted_right.csv",index=None)
```
